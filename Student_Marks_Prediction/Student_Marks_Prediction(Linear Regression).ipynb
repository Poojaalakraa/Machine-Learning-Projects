```python
'''Student Marks pridection'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
print("Required library is imported")
print("Data is imported")
df=pd.read_csv('/content/drive/MyDrive/Data set/Student_Marks .csv')
df


```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
user_input = int(input("Enter the number of hours studied: "))
predicted_marks = lr.predict([[user_input]])
print("Predicted marks for", user_input,"hours of study:",predicted_marks[0])
```

```python
plt.scatter(x_train,y_train,label='Actual Data',color='Orange')
plt.plot(x_train,lr.predict(x_train),color='yellow',label='Regression Line',linewidth=3)
plt.xlabel('Study_Hours')
plt.ylabel('Marks')
plt.title("Student Marks Prediction")
plt.legend()
plt.show()
```
