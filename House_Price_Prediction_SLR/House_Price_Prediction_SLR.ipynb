```python
''' House prediction based on area of house'''
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
df=pd.read_csv('/content/drive/MyDrive/Data set/House_price.csv')
df
```

```python
df.head()
print(df.dtypes)
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
area=float(input("Enter house area:-"))
predicted_price=lr.predict([[area]])
print("Predicted price for house area:",predicted_price[0],"Lakhs")
```

```python
plt.scatter(x_train,y_train,label='Actual Data',color='Orange')
plt.plot(x_train,lr.predict(x_train),color='yellow',label='Regression Line',linewidth=3)
plt.xlabel('Area')
plt.ylabel('Price in lakh')
plt.title("House price Prediction")
plt.legend()
plt.show()
```
